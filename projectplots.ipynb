```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("saved_models/output.csv", sep=";")
df.columns = ['Episode', 'AvgDrag', 'AvgLift', 'AvgRecircArea']


#
```

```python
print(df.columns)
```

```python
 # extract data
episodes = df['Episode']
avg_drag = df['AvgDrag']
avg_lift = df['AvgLift']
avg_recirc_area = df['AvgRecircArea']

# plot results
fig, ax = plt.subplots(nrows=3, figsize=(10, 8))
fig.suptitle('Model Performance', fontsize=14)

ax[0].plot(episodes, avg_drag, 'r', label='Average Drag')
ax[0].set_ylabel('Drag Coefficient')
ax[0].legend()

ax[1].plot(episodes, avg_lift, 'b', label='Average Lift')
ax[1].set_ylabel('Lift Coefficient')
ax[1].legend()

ax[2].plot(episodes, avg_recirc_area, 'g', label='Average Recirculation Area')
ax[2].set_xlabel('Episode')
ax[2].set_ylabel('Recirculation Area')
ax[2].legend()

plt.show()
```
